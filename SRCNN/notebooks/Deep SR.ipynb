```python
# main.py

import torch
from torch.utils.data import DataLoader
from torchvision import transforms
import sys

sys.path.append(r'C:\Users\dk412\Desktop\David\Python Projects\HyperSuperResolution\prod_code')
sys.path.append(r'C:\Users\dk412\Desktop\David\Python Projects\HyperSuperResolution\prod_code\inference.py')
from utils import SRCNN, Set14Dataset, train_model
from inference import *
```

Training

```python
def main():
    # Set device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    # Initialize model
    model = SRCNN().to(device)
    
    # Define transforms
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                           std=[0.229, 0.224, 0.225])
    ])
    
    image_size = (256, 256)
    
    # Create datasets with fixed image size
    train_dataset = Set14Dataset(
        root_dir=r'C:\Users\dk412\Desktop\David\Python Projects\HyperSuperResolution\data\Set14\Set14_train',
        scale_factor=2,
        transform=transform,
        image_size=image_size
    )
    
    val_dataset = Set14Dataset(
        root_dir=r'C:\Users\dk412\Desktop\David\Python Projects\HyperSuperResolution\data\Set14\Set14_val',
        scale_factor=2,
        transform=transform,
        image_size=image_size
    )
    
    print(f"Training dataset size: {len(train_dataset)}")
    print(f"Validation dataset size: {len(val_dataset)}")
    
    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)
    
    # Train model
    train_model(model, train_loader, val_loader, num_epochs=150, device=device)

if __name__ == '__main__':
    main()
```
